```python
import pandas as pd
import numpy as np
dat={'Temperature':['Hot','Cold','Very Hot','Hot','Warm', 'Warm', 'Hot', 'Hot', 'Cold','Very Hot'],
'Color':['Red','Yellow','Blue','Blue','Red','Yellow','Red','Yellow','Yellow','Yellow'],
'Target':[1,1,1,0,1,0,1,0,1,1]}
df=pd.DataFrame(dat,columns=['Temperature','Color','Target'])
df1=df
```

```python
df
```

```python
#One hot encoding
df1=pd.get_dummies(df,prefix=['Temp'],columns=['Temperature'])
df1
```

```python
from sklearn.preprocessing import OneHotEncoder
ohc=OneHotEncoder()
ohe=ohc.fit_transform(df.Temperature.values.reshape(-1,1)).toarray()
dfonehot=pd.DataFrame(ohe,columns=["Temp_"+str(ohc.categories_[0][i]) for i in range(len(ohc.categories_[0]))])
dfh=pd.concat([df,dfonehot],axis=1)
print(dfh)
```

```python
df=df1
ce_one_hot=ce.OneHotEncoder(cols="Temperature")
ce_one_hot.fit_transform(df)
```

```python
#Label Encoding
#In this encoding, each category is assigned a value from 1 through N
from sklearn.preprocessing import LabelEncoder
df['Temp_label_encoding']= LabelEncoder().fit_transform(df.Temperature)
df
```

```python
#Ordinal Encoding
#We do Ordinal encoding to ensure the encoding of variables retains the ordinal nature of the variable.

# Ordinal encoding will assign values as ( Cold(1) <Warm(2)<Hot(3)<â€Very Hot(4))

Temp_dict={'Cold':1, 'Warm':2, 'Hot':3, 'Very Hot':4}
df['Temp_Ordinal']=df.Temperature.map(Temp_dict)
df
```

```python
#Helmert Encoding
#In this encoding, the mean of the dependent variable for a level is compared to the mean of the dependent variable over all previous levels.
import category_encoders as ce
encoder=ce.HelmertEncoder(cols=['Temperature'],drop_invariant=True)
dfh=encoder.fit_transform(df['Temperature'])
df=pd.concat([df,dfh],axis=1)
```

```python
dat={'Temperature':['Hot','Cold','Very Hot','Hot','Warm', 'Warm', 'Hot', 'Hot', 'Cold','Very Hot'],
'Color':['Red','Yellow','Blue','Blue','Red','Yellow','Red','Yellow','Yellow','Yellow'],
'Target':[1,1,1,0,1,0,1,0,1,1]}
df=pd.DataFrame(dat,columns=['Temperature','Color','Target'])
df1=df
```

```python
#Binary encoder
import category_encoders as ce
encoder=ce.BinaryEncoder(cols=['Temperature'])
dfbin=encoder.fit_transform(df['Temperature'])
df=pd.concat([df,dfbin],axis=1)
df
```

```python
#Frequency encoding
#It is a way to utilize the frequency of the categories as labels. 
```

```python
df=df1
fr=df.groupby('Temperature').size()/len(df)
df.loc[:,'Temp_freq_encode']=df['Temperature'].map(fr)
df
```

```python
#Mean encoding
#Mean Encoding or Target Encoding is one viral encoding approach followed by Kagglers.
#Mean encoding is similar to label encoding, except here labels are correlated directly with the target.
```

```python
df=df1
mean_encode=df.groupby('Temperature')['Target'].mean()
print(mean_encode)
df.loc[:,'Temp_mean_encoding']=df['Temperature'].map(mean_encode)
df
```

```python
#BaseN Encoding
#When the BaseN base = 1 it is basically the same as one hot encoding. When base = 2 it is basically the same as binary encoding.
df=df1
ce_basen=ce.BaseNEncoder(cols=['Temperature'])
ce_basen.fit_transform(df)
```

```python
#Hashing encoder
df=df1
ce_basen=ce.HashingEncoder(cols=['Temperature'])
ce_basen.fit_transform(df)
```
